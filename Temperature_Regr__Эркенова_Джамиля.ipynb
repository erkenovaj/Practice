# Перевод градусов Цельсия в шкалу по Фаренгейту

Импортируем необходимые модули

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense

```

В качестве обучающего множества будем использовать два списка: температура по Цельсию и по Фаренгейту

```python
c = np.array([-40, -10, 0, 8, 15, 22, 38])
f = np.array([-40, 14, 32, 46, 59, 72, 100])
```

Определим модель НС. Добавим в эту модель слой нейронов, состоящий из одного выходного нейрона, имеющий ровно один вход и линейную активационную функцию.

Скомпилируем модель, указав критерий качества и способ оптимизации алгоритма градиентного спуска. В рамках данной задачи выберем минимум среднего квадрата ошибки и оптимизацию по Adam

```python
model = keras.Sequential()
model.add(Dense(units = 1, input_shape = (1,), activation = 'linear'))
model.compile(loss = 'mean_squared_error', optimizer = keras.optimizers.Adam(0.1))
```

Для запуска обучения используется метод fit

```python
history = model.fit(c, f, epochs = 1000)
```

Подадим на вход произвольное значение, нужно воспользоваться методом predict. Для отображения весовых коэффициентов, запишем метод get_weights


```python
print(model.predict([100]))
print(model.get_weights())
```

Обратимся к истории обучения. Возьмем значения критерия качества (функции потерь – loss function) и отобразим в виде графика

```python
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```

Скомпилируем модель, указав критерий качества и способ оптимизации алгоритма градиентного спуска. В рамках данной задачи выберем mean_squared_logarithmic_error и оптимизацию по SGD

```python
model = keras.Sequential()
model.add(Dense(units = 1, input_shape = (1,), activation = 'linear'))
model.compile(loss = 'mean_squared_logarithmic_error', optimizer = keras.optimizers.SGD(learning_rate=0.1))
```

```python
history = model.fit(c, f, epochs = 1000)
```

```python
print(model.predict([100]))
print(model.get_weights())
```

```python
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```

Уменьшим число эпох до 500

```python
history = model.fit(c, f, epochs = 500)
```

```python
print(model.predict([100]))
print(model.get_weights())
```

```python
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```

Построим модель с большим числом нейронов

```python
model = keras.Sequential()
model.add(Dense(units = 4, input_shape = (1,), activation = 'linear'))
model.compile(loss = 'mean_squared_error', optimizer = keras.optimizers.Adam(0.1))
```

```python
history = model.fit(c, f, epochs = 1000)
```

```python
print(model.predict([100]))
print(model.get_weights())
```
